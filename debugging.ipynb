```python
def create_knowledge_base(disease, symptoms):
    disease_knowledge = Or()

    # Create the set of symptoms and add it the knowledge base
    for symptom in symptoms:
        symptom_knowledge = And(*[Symbol(f'{symptom}') for symptom in symptoms])
        disease_knowledge.add(symptom_knowledge)

    # Create an Implication with the disease as the consequent
    implication = Implication(disease_knowledge, disease)

    return implication
```
